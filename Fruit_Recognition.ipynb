```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow
from tensorflow import keras
from keras.models import Model, Sequential
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten
from keras.utils import to_categorical
import os
```

```python
train_path = 'Dataset/train/'
test_path = 'Dataset/test/'

Xtrain = []
Xtest = []

Ytrain = []
Ytest = []

fruits = os.listdir(train_path)
for fruit in fruits:
    images = os.listdir(train_path+"/"+fruit)
    for img in images:
        image = cv2.imread(train_path+"/"+fruit+"/"+img)
        image = cv2.resize(image, (64,64), interpolation = cv2.INTER_AREA)
        Xtrain.append(image)
        Ytrain.append(fruit)

fruits = os.listdir(test_path)
for fruit in fruits:
    images = os.listdir(test_path+"/"+fruit)
    for img in images:
        image = cv2.imread(test_path+"/"+fruit+"/"+img)
        image = cv2.resize(image, (64,64), interpolation = cv2.INTER_AREA)
        Xtest.append(image)
        Ytest.append(fruit)
        
Xtrain = np.array(Xtrain)
Ytrain = np.array(Ytrain)
Xtest = np.array(Xtest)
Ytest = np.array(Ytest)
```

# Converting Image to GrayScale

```python
def grayScale(image):
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
    return gray
```

```python
for i in range(len(Xtrain)):
    Xtrain[i] = grayScale(Xtrain[i])
```

```python
for i in range(len(Xtest)):
    Xtest[i] = grayScale(Xtest[i])
```

# Normalizing the Image

```python
norm_trainX = Xtrain.astype('float32')/255.0
norm_testX = Xtest.astype('float32')/255.0
```

```python
plt.imshow(Xtest[8])
```

```python
plt.imshow(norm_testX[8])
```

```python
# one hot encode target values
classes = dict()
i = 0

for index in range(len(Ytest)):
    if Ytest[index] not in classes.keys():
        classes[Ytest[index]] = i
        i+=1
    Ytest[index] = classes[Ytest[index]]
    
for index in range(len(Ytrain)):
    Ytrain[index] = classes[Ytrain[index]]

Ytrain = to_categorical(Ytrain)
Ytest = to_categorical(Ytest)
```

# Convolutional Neural Network Model

```python
cnn = Sequential()

cnn.add(Conv2D(32, (3, 3), activation='relu', input_shape=((64,64) + (3,))))

cnn.add(MaxPooling2D((2, 2)))

cnn.add(Conv2D(32, (3, 3), activation='relu', input_shape=((64,64) + (3,))))

cnn.add(MaxPooling2D((2, 2)))

cnn.add(Flatten())

cnn.add(Dense(units=32, activation='relu')) #hidden layers

cnn.add(Dense(units=3, activation='softmax')) #output layer

# compile model
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

cnn.fit(x=norm_trainX, y=Ytrain, epochs=10)
```
